```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.imdb.com/chart/top").text
soup = BeautifulSoup( page ,'lxml')
```

```python
body = soup.find('tbody',class_='lister-list').find_all("tr")
```

```python
for movies in body :
    try:
        name = movies.find('td',class_='titleColumn').a.text
        date = movies.find("span",class_='secondaryInfo').text[1:-1]
        rating = movies.find('strong').text

    except Exception as e:
        print('failed',e)
        
    print(f'''
    Name ={name}
    Release Date ={date}
    Rating ={rating}''')       
```
