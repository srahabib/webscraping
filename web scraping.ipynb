```python
import bs4
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
```

```python
url = "https://www.newegg.com/Video-Cards-Video-Devices/Category/ID-38?Tpk=video%20cards"
```

```python
request = uReq(url)
```

```python
html = request.read()
```

```python
request.close()
```

```python
pagesoup = soup(html,"html.parser")
```

```python
containers = pagesoup.find_all("div", {"class": "item-container"})
```

```python
filename = "info.csv"
f = open(filename , 'w')
f.write("brand, name , shipping\n")
```

```python
for container in containers:
    
    if len(container.div.find_all("div", {"class": "item-branding"}))> 0:
    brand = container.div.div.a.img["title"]
    title_container = container.findAll("a", {"class":"item-title"}) 
    product_name =  title_container[0].text 
    shipping_container = container.findAll("li", {"class":"price-ship"})
    shipping = shipping_container[0].text.strip()
    
    print("\n")
    print("brand :" + brand)
    print("product_name :" + product_name)
    print("shipping :" + shipping)
    f.write(brand + "," + product_name.replace(",","|") +"," + shipping +"\n")   
```

```python
f.close()
```
