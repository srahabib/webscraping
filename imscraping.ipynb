```python
import requests
from bs4 import BeautifulSoup
import urllib.request
import io
from PIL import Image
import time
```

```python
url ="https://www.google.com/search?q=minecraft+characters&tbm=isch&chips=q:minecraft+charaters,g_1:transparent:mh_JUhgdGH0%3D&hl=en&sa=X&ved=2ahUKEwjkxqTx48D0AhUMUBoKHTI_CQMQ4lYoAHoECAEQEQ&biw=1519&bih=754"
```

```python
page = requests.get(url).text
soup = BeautifulSoup(page, "html.parser")
```

```python
urls = set()
for thumbnailimg in soup.find_all('img'):
  link = thumbnailimg.get('src')

  if link and "http" in link:
    print(link)
    urls.add(link)
```

```python
def download_image(download_path, url ,file_Name):
    try:
       image_content = requests.get(url).content
       image_file = io.BytesIO(image_content)
       image = Image.open(image_file)
       file_path = download_path +file_Name

       with open(file_path,"wb") as f:
          rgb_im = image.convert('RGB')
          rgb_im.save(f,"JPEG") 

       print("image downloaded")
    except Exception as e:
        print('Failed ',e)
```

```python
for i, url in enumerate(urls):
   download_image("G:\\imgs\\",url,str(i)+".jpg")
```
